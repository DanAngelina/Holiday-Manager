```python


def getHTML(url):
    response = requests.get(url)
    return response.text


html = getHTML("https://www.timeanddate.com/holidays/us/")

soup = BeautifulSoup(html,'html.parser')

table = soup.find('table')


data = []

for row in table.find('tbody'):
    holiday = {}
    row.find_all_next('a')
    holiday['name'] = row.find_all_next('a')[0].string
    row.find_all_next('th')
    holiday['date'] = row.find_all_next('th')[0].string
    data.append(holiday)

holidays = []

for i in data:
    if i not in holidays:
        holidays.append(i)
    
list(holidays)

with open("HolidaysUpdated.json", "w") as f:
    json.dump(holidays, f, indent=4)
```
